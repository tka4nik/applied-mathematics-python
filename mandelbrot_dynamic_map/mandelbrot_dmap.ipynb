```python
import holoviews as hv
from funcs import mandelbrot_par
hv.extension('bokeh')
```

```python
w, h = 800, 800
it = 5000
xmin, xmax = -1.5, 0.5
ymin, ymax = -1.0, 1.0
```

```python
def fractal(x_range, y_range):
    x0, x1 = x_range
    y0, y1 = y_range
    
    arr = mandelbrot_par(x0, x1, -y1, -y0, w, h, it)
    return hv.Image(arr, bounds=(x0, y0, x1, y1))
```

```python
range_xy = hv.streams.RangeXY(x_range=(xmin, xmax),
                              y_range=(ymin, ymax))

dmap = hv.DynamicMap(fractal, streams=[range_xy])
dmap.opts(hv.opts.Image(height=h, width=w, 
                        cmap='twilight', logz=True))
```
