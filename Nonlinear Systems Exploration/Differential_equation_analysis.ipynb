```python
import numpy as np
import matplotlib.pyplot as plt

r_range = np.arange(-0.25, 0.5, 0.0001)

def fixed_point_0(r):
    return np.sqrt(1 + np.sqrt(4 * r + 1)) / np.sqrt(2)

def fixed_point_1(r):
    return np.sqrt(1 - np.sqrt(4 * r + 1)) / np.sqrt(2)

def fixed_point_2(r):
    return (-1) * np.sqrt(1 + np.sqrt(4 * r + 1)) / np.sqrt(2)

def fixed_point_3(r):
    return (-1) * np.sqrt(1 - np.sqrt(4 * r + 1)) / np.sqrt(2)

def fixed_point_4(r):
    return r * 0

def map_derivative(x, r):
    return r + 3 * x ** 2 - 5 * x ** 4
```

```python
def get_fixed_point_data(point, r_array):
    data = np.array([point(r) for r in r_array])
    return r_array[np.invert(np.isnan(data))], data[np.invert(np.isnan(data))]


def plot_fixed_point_data_with_stability(r_array, y_array):
    mask = [map_derivative(y, x) < 0 for x,y in zip(r_array, y_array)]
    plt.plot(r_array[mask], y_array[mask], '-k')
    plt.plot(r_array[np.invert(mask)], y_array[np.invert(mask)], '--k')
```

```python
plt.figure(figsize=(5, 5), dpi=140)
plot_fixed_point_data_with_stability(*get_fixed_point_data(fixed_point_0, r_range))
plot_fixed_point_data_with_stability(*get_fixed_point_data(fixed_point_1, r_range))
plot_fixed_point_data_with_stability(*get_fixed_point_data(fixed_point_2, r_range))
plot_fixed_point_data_with_stability(*get_fixed_point_data(fixed_point_3, r_range))
plot_fixed_point_data_with_stability(*get_fixed_point_data(fixed_point_4, r_range))
plt.plot(0, 0, 'ko')
plt.plot(-0.25, 0.707, 'ko')
plt.plot(-0.25, -0.707, 'ko')
plt.title("Bifurcation diagram")
plt.legend(['stable', 'unstable'], loc="best")
plt.xlabel("r")
plt.ylabel("x")
```
